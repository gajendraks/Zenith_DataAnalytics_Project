```python
import pandas as pd 
import numpy as np
```

```python
df1=pd.read_csv("./../Data/pp_prob.csv")
df2=pd.read_csv("./../Data/batting_cluster.csv")
df3=pd.read_csv("./../Data/bowling_cluster.csv")
```

```python
df1=df1.round(2)
df1.head()
```

```python
df2.rename(columns = {"player_name":"batsman","cluster":"batsman_cluster"},inplace=True)
df3.rename(columns = {"player_name":"bowler","cluster":"bowler_cluster"},inplace=True)
```

```python
df2.head()
```

```python
df1=df1.merge(df2,how='left',left_on=["batsman"],right_on=["batsman"])
df1=df1.merge(df3,how='left',left_on=["bowler"],right_on=["bowler"])
```

```python
df1.head()
```

```python
df1.drop(columns=['batsman','bowler'])
clusters_prob = df1.groupby(["batsman_cluster","bowler_cluster"])['P0','P1','P2','P3','P4','P6','p_not_out'].mean().round(2).reset_index()
```

```python
clusters_prob[['batsman_cluster','bowler_cluster']]=clusters_prob[['batsman_cluster','bowler_cluster']].astype(int)
```

```python
clusters_prob.to_csv("./../Data/cluster_prob.csv",index=None)
```
